```
from typing import Dict
import string


def analyze_text(data: str) -> Dict[str, float | int]:
    """
    Анализирует текстовую строку и возвращает статистику по словам.

    Функция очищает текст от знаков пунктуации, приводит его к нижнему
    регистру,разбивает на слова:
    - общее количество слов
    - количество уникальных слов
    - среднюю длину слова

    Args:
        data (str): Входная текстовая строка для анализа

    Returns:
        Dict[str, float | int]: Словарь со статистикой:
            - 'total_words': общее количество слов (int)
            - 'unique_words': количество уникальных слов (int)
            - 'avg_word_length': средняя длина слова (float)
    """
    data = data.lower()

    # Удаляем знаки пунктуации
    translator = str.maketrans('', '', string.punctuation)
    data = data.translate(translator)

    # Разбиваем текст на слова
    words = data.split()

    # Находим уникальные слова
    unique_words = set(words)

    if words:  # Проверяем, что список не пустой
        avg_length = sum(len(word) for word in words) / len(words)
    else:
        avg_length = 0.0

    stats = {
        'total_words': len(words),
        'unique_words': len(unique_words),
        'avg_word_length': avg_length
    }

    return stats


raw_text = "Hello, world! Hello everyone. This is a test... A TEST!"
stats = analyze_text(raw_text)
print(stats)
```
